# WeatherPy

* Initial imports and setup

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
import scipy.stats as st
from scipy.stats import linregress

# Import API key & Citipy to determine city based on lat & long
from api_keys import weather_api_key
from citipy import citipy

# Create Output File (CSV)
output_data_file = "../output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# Create empty lists to house Lat & Longs, as well as cities. 
lat_lngs = []
cities = []

# Random latitude and longitude combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)
#Verify Count
len(cities)
```

```python
#Create empty lists to store
country_list = []
city_names_list = []
cloudiness_list = []
lat_list = []
lng_list = []
dates_list = []
humidity_list = []
max_temp_list = []
wind_speed_list = []
index_counter = 0
set_counter = 1
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
print("Data Retrieval ")
print("___________________________")

base_url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"
query_url = f"{base_url}appid={weather_api_key}&units={units}&q="


#Do the following for each city in cities list
for index, city in enumerate(cities, start = 1):
    
#Use exception to avoid errors from preventing the script to run. Run it through.  
    try:
        response = requests.get(query_url + city).json()
        city_names_list.append(response["name"])
        cloudiness_list.append(response["clouds"]["all"])
        country_list.append(response["sys"]["country"])
        dates_list.append(response["dt"])
        humidity_list.append(response["main"]["humidity"])
        lat_list.append(response["coord"]["lat"])
        lng_list.append(response["coord"]["lon"])
        max_temp_list.append(response['main']['temp_max'])
        wind_speed_list.append(response["wind"]["speed"])
        
        
        if index_counter > 49:
            index_counter = 0
            set_counter = set_counter + 1
    
        else:
            index_counter = index_counter + 1   
        print(f"Processing Record {index_counter} of Set {set_counter} : {city}") 
  

    except(KeyError, IndexError):
        print("City not found. Skipping...")

print("___________________________")
print("Retrieval Complete")
print("___________________________")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#Use Pandas to create a data frame using the data that was just gathered.
weather_dict = pd.DataFrame({ 
                "Country" : country_list,
                "City" : city_names_list,
                "Lat" : lat_list,
                "Lng" : lng_list,
                "Cloudiness" : cloudiness_list,
                "Date" : dates_list,
                "Humidity" : humidity_list,
                "Max Temp" : max_temp_list,
                "Wind Speed" : wind_speed_list
})

weather_dict.count()
```

```python
#Save data to output file
weather_dict.to_csv("../output_data/cities.csv", index = False)
```

## Linear Regression

```python
northern_hemi = weather_dict.loc[weather_dict["Lat"] >= 0]
southern_hemi = weather_dict.loc[weather_dict["Lat"] < 0]
```

```python
def linear_regression(x,y):
    print(f"The r-squared is : {round(st.pearsonr(x, y)[0],2)}")
    (slope, intercept, rvalue, pvalue, stderr) = linregress(x, y)
    regress_values = x * slope + intercept
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
    plt.scatter(x, y)
    plt.plot(x,regress_values,"r-")
    return line_eq

#Annotate equation
def annotate(line_eq, a, b):
    plt.annotate(line_eq,(a,b),fontsize=15,color="red")
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
#Linear Regression
equation = linear_regression(northern_hemi["Lat"], northern_hemi["Max Temp"])
annotate(equation, 5, 10)

#Set titles and labels
plt.title("Northern Hemisphere - Max Temp vs. Latitude Linear Regression")
plt.xlabel("Latitude")
plt.ylabel("Max Temp (F)")

#Save as png
plt.savefig("../Images/Northern Hemisphere - Max Temp vs. Latitude Linear Regression.png")
```
